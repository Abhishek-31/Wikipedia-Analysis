```python
import re
import wikipedia
import wikipediaapi
```

```python
with open("/home/abhishek/Documents/Wikipedia-Analysis/FAQ-Builder/City Data collector/final2.txt","r") as f:
    z=f.read().split('\n')
myset=set(z)
z=list(myset)
print(len(z))
print(z)
with open("final2.txt","w+") as f:
    for _ in z:
        f.write(_)
        f.write("\n")
print(z)
```

```python
a={}
c=0
print("Files done")
wiki = wikipediaapi.Wikipedia('en')
for _ in z:
    print(">>")
    try:
        content=wikipedia.page(_).content
        try:
            k=content.find("== See also ==")
        except ValueError as e:
            print("solving *")
            try:
                k=content.find("== References ==")
            except:
                try:
                    k=content.find("== Further reading ==")
                except:
                    try:
                        k=content.find("== External links ==")
                    except:
                        pass
                    else:
                        content=content[0:k]
                else:
                    content=content[0:k]
                    
            else:
                content=content[0:k]
                
        else:
            content=content[0:k]
        
            
        if(c==0):
            f=open('%s.txt' % _,'w+')
            f.write(content)
            f.write("\n\n\n\n\n")
            f.write("tags\n")
            string=' '
            tags=re.findall(r'== (.*?) ==',content)
            
            if("References" in tags):
                tags.remove("References")
            if("External links" in tags):
                tags.remove("External links")
            if("See also" in tags):
                tags.remove("See also")
            if("Further reading" in tags):
                tags.remove("Further reading")
            l=string.join(tags)
            f.write(l)
        c=c+1
        a[_]={}
        a[_]["content"]=content
        tags=re.findall(r'== (.*?) ==',content)
        if("References" in tags):
            tags.remove("References")
        if("External links" in tags):
            tags.remove("External links")
        if("See also" in tags):
            tags.remove("See also")
        if("Further reading" in tags):
            tags.remove("Further reading")
        a[_]["headings"]= tags
        print(c)
        
    except Exception as exception:
        print("Inside an exception")
        print(type(exception))
        if(type(exception)==wikipedia.exceptions.DisambiguationError):
            print("Error coming for",end=" -> ")
            print(_)
            for check in exception.options:
                pg=wiki.page(check)
                if(pg.exists()):
                    c=c+1
                    print(c)
                    print(pg)
                    try:    
                        content=wikipedia.page(check).content
                        try:
                            x=content.find(".")
                        except:
                            print("rejected")
                        else:
                            l=content[0:x]
                            if("city" in l or "municipality" in l or "district" in l or "province" in l or "town" in l or "metropolis" in l):
                                print("accepted")
                                a[check]={}
                                try:
                                    k=content.find("== See also ==")
                                except:
                                    try:
                                        k=content.find("== References ==")
                                    except:
                                        try:
                                            k=content.find("== Further reading ==")
                                        except:
                                            try:
                                                k=content.find("== External links ==")
                                            except:
                                                pass
                                            else:
                                                content=content[0:k]
                                        else:
                                            content=content[0:k]
                    
                                    else:
                                        content=content[0:k]
                    
                                else:
                                    content=content[0:k]
                                a[check]["content"]=content
                                tags=re.findall(r'== (.*?) ==',content)
                                if("References" in tags):
                                    tags.remove("References")
                                if("External links" in tags):
                                    tags.remove("External links")
                                if("See also" in tags):
                                    tags.remove("See also")
                                if("Further reading" in tags):
                                    tags.remove("Further reading")
                                a[check]["headings"]=tags
                            else:
                                print("rejected")
                    except:`
                        pass
                else:
                    continue
        else:
            pass
```
