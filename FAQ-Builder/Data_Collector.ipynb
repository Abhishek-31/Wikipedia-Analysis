```python
import re
import wikipedia
import wikipediaapi
```

```python
with open("final.txt","r") as f:
    z=f.read().split('\n')
    print(z)
```

```python
a={}
c=0
print("Files done")
wiki = wikipediaapi.Wikipedia('en')
for _ in z:
    print(">>")
    try:
        content=wikipedia.page(_).content
        if(c==0):
            f=open('%s.txt' % _,'w+')
            f.write(content)
            f.write("\n\n\n\n\n")
            f.write("tags\n")
            string=' '
            tags=re.findall(r'== (.*?) ==',content)
            
            if("References" in tags):
                tags.remove("References")
            if("External links" in tags):
                tags.remove("External links")
            l=string.join(tags)
            print(l)
            f.write(l)
        c=c+1
        a[_]={}
        a[_]["content"]=content
        tags=re.findall(r'== (.*?) ==',content)
        if("References" in tags):
            tags.remove("References")
        if("External links" in tags):
            tags.remove("External links")
        a[_]["headings"]= tags
        print(c)
        
    except Exception as exception:
        print("Inside an exception")
        print(type(exception))
        if(type(exception)==wikipedia.exceptions.DisambiguationError):
            print(_)
            for check in exception.options:
                print(check)
                c=c+1
                pg=wiki.page(check)
                if(pg.exists()):
                    print(pg)
                    try:
                        content=wikipedia.page(check).content
                        a[check]={}
                        a[check]["content"]=content
                        tags=re.findall(r'== (.*?) ==',content)
                        if("References" in tags):
                            tags.remove("References")
                        if("External links" in tags):
                            tags.remove("External links")
                        a[check]["headings"]=tags 
                    except:
                        pass
                else:
                    continue
        else:
            pass
```
